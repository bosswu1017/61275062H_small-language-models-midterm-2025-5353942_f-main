````python
#my integer answer
def create_random_sequence_linear_with_outliers(a=None, b=None, p=None, l=None):
    "returns a sequences of length l in the form y = ax+b, with a,b being integers selected from -5 to 5 and p being 1" 

    # Step 1: Set default values if not provided
    if a is None:
        a = random.randint(-5, 5)
    if b is None:
        b = random.randint(-5, 5)
    if l is None:
        l = 10  # default length
    if p is None:
        p = 1   # default outlier count

    # Step 2: Generate sequence y = ax + b
    x_vals = list(range(l))
    seq = [a * x + b for x in x_vals]

    # Step 3: Choose p unique indices for outliers
    outliers = random.sample(range(l), p)

    # Step 4: Inject outliers (replace value with something extreme)
    for idx in outliers:
        seq[idx] = seq[idx] + random.choice([-100, 100])  # you can adjust magnitude

    # Step 5: Return results
    return seq, outliers

#my float answer
def create_random_sequence_linear_with_outliers_float(a=None, b=None, p=None, l=None):
    """
    Returns a sequence of length l in the form y = ax + b, with a,b as floats from -5.0 to 5.0.
    Replaces p values with large-deviation outliers and returns the sequence and outlier indices.
    """
    if a is None:
        a = random.uniform(-5.0, 5.0)
    if b is None:
        b = random.uniform(-5.0, 5.0)
    if l is None:
        l = 10
    if p is None:
        p = 1

    x_vals = list(range(l))
    seq = [round(a * x + b, 2) for x in x_vals]

    outliers = random.sample(range(l), p)
    for idx in outliers:
        seq[idx] = round(seq[idx] + random.choice([-100.0, 100.0]), 2)

    return seq, outliers


#test integer answer
seq, outliers = create_random_sequence_linear_with_outliers(a=2, b=1, p=3, l=50)
print("Generated sequence with outliers:", seq)
print("Outlier indices:", outliers)

#test float answer
seq, outliers = create_random_sequence_linear_with_outliers_float(a=1.5, b=0.3, p=2, l=10)
print("Float sequence with outliers:", seq)
print("Outlier indices:", outliers)
````
