<p>This equation models a system under constant acceleration (where d is the change in velocity (dv/dt), in other words the acceleration).</p>

````python

def create_random_sequence_quadratic(a=None, b=None, d=None, l=None):
    "returns a sequences of length l in the form y = ax+b, with a,b, and d being integers selected uniformly from -5 to 5" 
    """
    Generates a quadratic-like sequence with length l,
    where the difference between steps increases linearly (constant acceleration).
    """
    if a is None:
        a = random.randint(-5, 5)
    if b is None:
        b = random.randint(-5, 5)
    if d is None:
        d = random.randint(-2, 2)  # acceleration (變化量的變化量)
    if l is None:
        l = 10

    seq = []
    y = a  # 初始值
    delta = b  # 初始變化量（"速度"）
    
    for _ in range(l):
        seq.append(y)
        y += delta
        delta += d  # 每一步改變速度，相當於加速度

    return seq

def create_random_sequence_quadratic_float(a=None, b=None, d=None, l=None):
    """
    Generates a float quadratic-like sequence with constant acceleration,
    where the difference between values increases by d at each step.
    """
    if a is None:
        a = random.uniform(-5.0, 5.0)
    if b is None:
        b = random.uniform(-5.0, 5.0)
    if d is None:
        d = random.uniform(-1.0, 1.0)
    if l is None:
        l = 10

    seq = []
    y = a
    delta = b

    for _ in range(l):
        seq.append(round(y, 2))
        y += delta
        delta += d

    return seq

seq = create_random_sequence_quadratic(a=0, b=1, d=1, l=10)
print(seq)
print(create_random_sequence_quadratic_float())
````

This function generates a quadratic sequence by simulating constant acceleration: it starts from an initial value and applies an increment that grows by a fixed value d at every step.
