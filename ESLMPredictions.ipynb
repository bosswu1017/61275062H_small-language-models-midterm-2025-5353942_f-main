````python
#my integer answer
def create_random_sequence(a=None, b=None, l=None):
    "returns a sequences of length l in the form y = ax+b, with a and b being integers selected uniformly from -5 to 5" 
    """
    "Returns a sequence of length l in the form y = ax + b,
    with a and b being integers selected from -5 to 5 if not provided.
    """
    if a is None:
        a = random.randint(-5, 5)
    if b is None:
        b = random.randint(-5, 5)
    if l is None:
        l = 10  # default length
    
    seq = [a * x + b for x in range(l)]
    return seq
#my float answer
def create_random_sequence_float(a=None, b=None, l=None):
    """
    Returns a sequence of length l in the form y = ax + b,
    where a and b are floats randomly selected from -5.0 to 5.0 if not provided.
    """
    if a is None:
        a = random.uniform(-5.0, 5.0)
    if b is None:
        b = random.uniform(-5.0, 5.0)
    if l is None:
        l = 10
    
    seq = [round(a * x + b, 2) for x in range(l)]
    return seq

#test integer answer
print(create_random_sequence())  

#test float answer
print(create_random_sequence_float())
````
